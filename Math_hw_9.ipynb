```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```


Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

Экспоненциальное распределение

```python
rv     = sts.expon(0.5)        
sample = rv.rvs(size=1000) 
```

```python
x   = np.linspace(0,1,300)
pdf = rv.pdf(x)

plt.hist(sample, density=True, bins=20, label='sample histogram')
plt.plot(x,      pdf,  label='theoretical pdf', alpha=2)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
## параметры запуска
n             = [5, 10, 100, 300, 1000]  
samples_count = 1000                   

a             = np.zeros((len(n), samples_count))

for i in range(len(n)):                 
    for number in range(samples_count): 
        sample       = rv.rvs(size=n[i])
        a[i][number] = sample.mean()    
```

```python
fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(nrows=5, ncols=1, figsize=(10, 15))
labels = []

for i in n:
  labels.append('sample of ' + str(i) + ' histogram')

ax0.hist(a[0],  bins = 15,  label=labels[0])
ax1.hist(a[1],  bins = 15,  label=labels[1])
ax2.hist(a[2],  bins = 15,  label=labels[2])
ax3.hist(a[3],  bins = 15,  label=labels[3])
ax4.hist(a[4],  bins = 15,  label=labels[4])

ax0.set_ylabel('$f(x)$')
ax0.set_xlabel('$x$')
ax0.legend(loc='upper right')

ax1.set_ylabel('$f(x)$')
ax1.set_xlabel('$x$')
ax1.legend(loc='upper right')

ax2.set_ylabel('$f(x)$')
ax2.set_xlabel('$x$')
ax2.legend(loc='upper right')

ax3.set_ylabel('$f(x)$')
ax3.set_xlabel('$x$')
ax3.legend(loc='upper right')

ax4.set_ylabel('$f(x)$')
ax4.set_xlabel('$x$')
ax4.legend(loc='upper right')
```
