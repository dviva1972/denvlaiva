Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
•p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит; 
•s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится
Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.
•l – list – команда, которая выведет список всех документов в формате
passport "2207 876234" "Василий Гупкин" 
•as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
•d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ; 
•m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку; 
•a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.

```python
documents = [ 
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, 
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, 
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} 
]

directories = { '1': ['2207 876234', '11-2'], '2': ['10006'], '3': [] }

command_enter = input('Введите команду:') 
command_type=['p', 's', 'l', 'as', 'd', 'm', 'a']

def print_name(doc_data): 
    i=0 
    doc_number = input('ведите номер документа ') 
    for row in doc_data: 
        if row ['number'] == doc_number: 
            print(row['name']) 
            i+=1 
    if i==0: 
        print('Указан неверный номер документа')

def find_shelf (direct_data): 
    i=0 
    doc_number = input('ведите номер документа ') 
    for row in direct_data.items(): 
        if doc_number in row [1]: 
            print('Документ находится на полке № ', row [0]) 
            i+=1 
    if i==0: 
        print('Указан неверный номер документа')

def all_docs (doc_data): 
    for row in doc_data: 
        print('{a:10} "{b:11}"  "{c:18}"'.format(a=row['type'], b=row['number'], c=row['name']))
              
def new_shelf (direct_data): 
    number_new_shelf = input('ведите код новой полки ') 
    direct_data.setdefault(number_new_shelf, []) 
    print(direct_data)

def del_doc (doc_data, direct_data): 
    i=0 
    del_number = input('ведите номер документа для удаления ') 
    for row in doc_data:
        if del_number == row['number']: 
            doc_data.remove(row) 
            i+=1 
    {row1.remove(del_number) for row1 in direct_data.values() if del_number in row1}
    if i==0: 
        print('Удаление невозможно, введен несуществующий документ') 
    print(doc_data) 
    print(direct_data) 

def move_doc(direct_data):            
    doc_number = input('ведите номер документа ')
    shelf = input('ведите номер полки ')
    i=0
    if shelf not in direct_data.keys():
        print('Введенной полки не существует') 
    else:    
        for row in direct_data.values(): 
            if doc_number in row: 
                row.remove(doc_number) 
                i+=1
        if i==0:
            print ('Введенного документа не существует') 
        else:    
            for row1 in direct_data.items(): 
                if shelf ==row1[0]: 
                    row1[1].append(doc_number) 
    print(direct_data)

def add_doc(direct_data, doc_data):
    i=0
    number_doc = input('Введите номер документа ')
    shelf = input('Введите номер полки ') 
    sort_doc = input('Введите тип документа ') 
    name_doc = input('ведите имя владельца документа ')
    doc_form = ['type', 'number', 'name']
    for row in direct_data.items(): 
        if shelf ==row[0]: 
            doc_data.append(dict(zip(doc_form, [sort_doc, number_doc, name_doc])))
            row[1].append([number_doc]) 
            i+=1
    if i==0: 
        print('Введенной полки не существует') 
    print(doc_data)
    print(direct_data)

if command_enter not in command_type: 
    print('Вы ввели неправильную команду!') 
elif command_enter == 'p': 
    print_name (documents) 
elif command_enter == 's': 
    find_shelf (directories) 
elif command_enter == 'l': 
    all_docs(documents) 
elif command_enter == 'as': 
    new_shelf(directories) 
elif command_enter == 'd': 
    del_doc(documents, directories) 
elif command_enter == 'm': 
    move_doc(directories)
elif command_enter == 'a': 
    add_doc(directories, documents)
```
