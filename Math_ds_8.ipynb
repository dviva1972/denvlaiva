Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

Нормальное распределение

```python
x = np.linspace(-10, 10, 1000)

norm_rv = sts.norm(loc = 0, scale = 2)
norm_1 = norm_rv.pdf(x)
plt.plot(x, norm_1)

norm_rv = sts.norm(loc = -3, scale = 2)
norm_2 = norm_rv.pdf(x)
plt.plot(x, norm_2)

norm_rv = sts.norm(loc = 3, scale = 2)
norm_3 = norm_rv.pdf(x)
plt.plot(x, norm_3)

norm_rv = sts.norm(loc = 0, scale = 1)
norm_4 = norm_rv.pdf(x)
plt.plot(x, norm_4)

norm_rv = sts.norm(loc = 0, scale = 8)
norm_5 = norm_rv.pdf(x)
plt.plot(x, norm_5)


plt.legend(['loc=0,  scale=2', 
            'loc=-3, scale=2', 
            'loc=3,  scale=2', 
            'loc=0,  scale=1', 
            'loc=0,  scale=8'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.grid()
plt.show()
```

В нормальном распределение математическое ожидание определяет сдвиг графика по оси X 

СКО определяет меру отличия экстремума от среднего значения Y

Экспоненциальное распределение:


```python
x = np.linspace(0,10,1000)

expon_rv = sts.expon(0.1)
expon_1 = expon_rv.pdf(x)
plt.plot(x, expon_1)

expon_rv = sts.expon(0.5)
expon_2 = expon_rv.pdf(x)
plt.plot(x, expon_2)

expon_rv = sts.expon(1)
expon_3 = expon_rv.pdf(x)
plt.plot(x, expon_3)

expon_rv = sts.expon(2)
expon_4 = expon_rv.pdf(x)
plt.plot(x, expon_4)

expon_rv = sts.expon(5)
expon_5 = expon_rv.pdf(x)
plt.plot(x, expon_5)

plt.legend(['exp=0.1', 'exp=0.5', 'exp=1', 'exp=2', 'exp=5'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.grid()
plt.show()
```

Параметр лямба влияет на смещение графика по оси X


Распределение Стьюдента:

```python
x = np.linspace(-5, 5, 1000)

t_rv = sts.t(0.001)
t_1 = t_rv.pdf(x)
plt.plot(x, t_1)

t_rv = sts.t(0.1)
t_2 = t_rv.pdf(x)
plt.plot(x, t_2)

t_rv = sts.t(1)
t_3 = t_rv.pdf(x)
plt.plot(x, t_3)

t_rv = sts.t(10)
t_4 = t_rv.pdf(x)
plt.plot(x, t_4)

t_rv = sts.t(100000)
t_5 = t_rv.pdf(x)
plt.plot(x, t_5)

plt.legend(['t 0.001', 't 0.1', 't 1', 't 10', 't 100000'])
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.grid()
plt.show()
```


Число степеней свободы влияет на ширину графика плотности свободы, чем больше его значение, тем ближе распределение Стьюдента стремится к нормальному.

Задание 2

Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания $\xi$ в отрезок $[3.5;7]$


Распределение равномерное на интеравале от $a$ до $b$, где $a=0$, $b=5$.

Плотность распределения равна: $C = \frac {1}{5-0} = 0.2$

Математическое ожидание - середина интервала, т.е. $\mu = 2.5$

Вероятность попасть в отрезок $[3.5;7]$ равна вероятности попасть в отрезок $[3.5;5]$, так как вероятность попасть в интервал $(5;7]$ равна 0.

Вероятность попасть в отрезок от $[3.5;5]$ пропорциональна длине этого отрезка и вычисляется, как: $\frac {5-3.5}{5-0} = 0.3$

Задание 3

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

Полная продолжительность работы светофора равна: $T = 54 + 4 + 28 = 86 (с)$

Найдем вероятность того, что светофор горит красным, желтым, зеленым:

$P(R) = \frac {54}{86} = \frac {27}{43} $,

$P(Y) = \frac {4}{86} = \frac {2}{43} $,

$P(G) = \frac {28}{86} = \frac {14}{43} $.

Количество информации можно посчитать по формуле Шэннона:
$H(x) = -\sum\limits_{i=1}^n p_i \cdot log_2 p_i$

```python
 - (27/43 * math.log2(27/43) + 2/43 * math.log2(2/43) + 14/43 * math.log2(14/43))
```

сообщение о цвете светофора в текущий момент несет в себе количество информации, равное 2 бит.

В случае одинаковой продолжительности для каждого цвета вероятность равна:

$P(R) = P(Y) = P(G) = \frac {28}{84} = \frac {1}{3} $

Тогда по формуле Шэннона:

```python
- (1/3 * math.log2(1/3) * 3)
```
