```python
import numpy as np
from scipy.misc import derivative 
import matplotlib.pyplot as plt
```

##### 1.  посчитайте значение 1-ой и 2-ой производной фунции:   f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

1.1. Scipy  - по простому

```python
def f(x):
    return x ** 5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)

x = 1

f1_x = derivative(f, x, dx=1e-6, n=1)
f2_x = derivative(f, x, dx=1e-6, n=2)

print('Значение 1-ой производной функции  точке  x = ', f1_x.round(3))
print('Значение 2-ой производной функции  точке  x = ', f2_x.round(3))
```

1.2. Numpy -  менее по простому

```python
def diff_np_1(x): 
    h = 1e-6 
    return (f(x+h)-f(x-h))/(2*h) 
```

```python
diff_np_1(1)
```

```python
def diff_np_2(x): 
    h = 1e-5 
    return (diff_np_1(x+h)-diff_np_1(x-h))/(2*h) 
```

```python
diff_np_2(1)
```

##### 2.  Постройте график первой и второй производной функции  (sin(2x+1))^5   на отрезке [-5; 5]

```python
def f(x):
    return (np.sin(2 * x + 1)) ** 5

interval = np.linspace(-5, 5)

fx = f(x)
f1x = [derivative(f, x, dx=1e-6, n=1) for x in interval]
f2x = [derivative(f, x, dx=1e-6, n=2) for x in interval]

plt.plot(x, fx,   linestyle = '-',  linewidth = 3, color = 'r')
plt.plot(x, f1x,  linestyle = '--', linewidth = 2, color = 'g')
plt.plot(x, f2x,  linestyle = '-.', linewidth = 1, color = 'b')

plt.title("График функции и ее производных")
plt.legend(["f (x)", "f '(x)", "f ''(x)"])
plt.show()
```
